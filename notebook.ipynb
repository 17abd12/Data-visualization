```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import hvplot.pandas
import holoviews as hv
import numpy as np
```

Reading csv in using pd.read_csv and than creating a data frame and than merging those dataframes to in a summary_df data frame. Now till now we have two dataframes one is of summary and other is of cinema_vistors that are going to be furthur used

```python
cinema_age = pd.read_csv('data2/CinemaAge.csv', index_col=0)
cinema_Overhead = pd.read_csv('data2/CinemaOverheads.csv', index_col=0)
cinema_Capacity = pd.read_csv('data2/CinemaCapacity.csv', index_col=0)
cinema_marketing = pd.read_csv('data2/CinemaMarketing.csv', index_col=0)
cinema_spend = pd.read_csv('data2/CinemaSpend.csv', index_col=0)
cinema_Visitors = pd.read_csv('data2/CinemaWeeklyVisitors.csv', index_col=0)
cinema_age=cinema_age.rename(columns={'Avg age (yrs)':'Age'})
cinema_Capacity=cinema_Capacity.rename(columns={'Seating capacity':'Capacity'})
cinema_Overhead=cinema_Overhead.rename(columns={'Overheads (£000s)':'Overhead'})
cinema_marketing=cinema_marketing.rename(columns={'Marketing (£000s)':'Marketing'})
cinema_spend=cinema_spend.rename(columns={'Avg spend (£)':'spend'})
Summary_df = pd.merge(cinema_Overhead, cinema_Capacity,on='Id')
Summary_df = pd.merge(Summary_df, cinema_spend,on='Id')
Summary_df = pd.merge(Summary_df, cinema_age,on='Id')
Summary_df = pd.merge(Summary_df, cinema_marketing,on='Id')
cust_series = cinema_Visitors.sum()
Summary_df['Anual Customer'] = cust_series
```

Line plots with 6-months rolling average and trendlines large conemas. These plots will highlight the average customer frequency over halfyear periods for just large cinemas. 

```python


# Register converters for matplotlib
pd.plotting.register_matplotlib_converters()

period = 32
rolling_average = cinema_Visitors.rolling(window=period).mean()

plt.figure(figsize=(14, 14))
plt.plot(cinema_Visitors[mid_visitors_df.index], linewidth=0.4)
selected = mid_visitors_df.index
plt.gca().set_prop_cycle(None)

for name in selected:
    x = np.arange(len(cinema_Visitors[name]))
    z = np.polyfit(x, cinema_Visitors[name], 1)
    trend = np.poly1d(z)
    plt.plot(cinema_Visitors.index, trend(x), linestyle='--')

plt.gca().set_prop_cycle(None)
plt.plot(rolling_average[selected], linewidth=1.5)

# Set x-axis tick locations and labels
plt.xticks(rotation=45)
plt.gca().xaxis.set_major_locator(plt.MaxNLocator(10))

plt.xlabel('Date', fontsize=18)
plt.ylabel('Number of Customers', fontsize=18)
plt.show()
```

This code creates a figure with two subplots. It then plots the data for large cinemas in the first subplot and medium cinemas in the second subplot. For each cinema, it calculates the condensed series using the condense_series() function, performs linear regression to obtain the trendline, and plots the data points and trendlin

```python
import matplotlib.dates as mdates
def condense_series(series):
    condensed_series = pd.Series()
    for i in range(0, len(series), 20):
        key = pd.to_datetime(series.index[i]).strftime("%b-%Y")  
        value_sum = series.iloc[i:i+5].sum()
        second_value = series.iloc[i+1]
        condensed_series[key] = value_sum + second_value
    return condensed_series
fig, (ax1, ax2) = plt.subplots(2, 1, sharex=True, figsize=(10, 10))  
lines1 = []
labels1 = [] 
for name in large_visitors_df.index:
    data_points = condense_series(cinema_Visitors[name])
    x = np.arange(len(data_points))
    y = np.array(list(data_points.values)).astype(np.float64)
    z = np.polyfit(x, y, 1)
    trendline = np.poly1d(z)
    ax1.plot(x, y, label=name)
    line, = ax1.plot(x, trendline(x), linestyle='--', linewidth=2)
    lines1.append(line)
    labels1.append(name)
ax1.set_title('14-Months Rolling Averages & Trendlines of large Cinemas')
ax1.set_ylabel('Number of Customers')
lines2 = []
labels2 = []
for name in mid_visitors_df.index:
    data_points = condense_series(cinema_Visitors[name])
    x = np.arange(len(data_points))
    y = np.array(list(data_points.values)).astype(np.float64)
    z = np.polyfit(x, y, 1)
    trendline = np.poly1d(z)
    ax2.plot(x, y, label=name)
    line, = ax2.plot(x, trendline(x), linestyle='--', linewidth=2)
    lines2.append(line)
    labels2.append(name)
ax2.set_title('14-Months Rolling Averages & Trendlines of medium Cinemas')
ax2.set_ylabel('Number of Customers')
ax1.set_xticks(np.arange(len(data_points)))
ax1.set_xticklabels(list(data_points.keys()), rotation=45, ha='right')
ax1.xaxis.set_major_locator(mdates.AutoDateLocator())
ax2.set_xticks(np.arange(len(data_points)))
ax2.set_xticklabels(list(data_points.keys()), rotation=45, ha='right')
ax2.xaxis.set_major_locator(mdates.AutoDateLocator())
fig.legend(lines1 + lines2, labels1 + labels2) 
plt.tight_layout()
plt.show()
```
